```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                     'key2' : [4,5,6,7,8],
                     'key3' : [3,4,5,6,6]})
```

```python
data1
```

```python
data2 = pd.DataFrame({'key1' : [1,2,45,6,67],
                     'key4' : [56,5,6,7,8],
                     'key5' : [3,56,5,6,6]})
```

```python
data2
```

```python
pd.merge(data1,data2)
```

```python
pd.merge(data2,data1)
```

```python
pd.merge(data1,data2,how='left')
# how = 'left' keeps all the data in data1 as it is and match the data in data1 and data2 and wherever the data matches or intersects it pri
# prints the value ,if it doesn't match it will return NaN value for that
```

```python
pd.merge(data1,data2,how='right')
# returns all the data in right dataframe i.e. in 'data2' here as it is,but where the data intersects in data2 from data1 it returns that va
# value as it is in floating point number but all the remained values that are not matching it returns the NaN value there in a dataframe.
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='cross')
# combines to create a combination of all possible outcomes
# first it will take first row of 'data1' and merge it with all rows in 'data2' and process it further for all rows in 'data1'
```

```python
pd.merge(data1,data2,how='inner') # by default --> how = 'inner'
```

 > how parameters -->
 
     * left: use only keys from left frame, similar to a SQL left outer join;
      preserve key order.
    * right: use only keys from right frame, similar to a SQL right outer join;
      preserve key order.
    * outer: use union of keys from both frames, similar to a SQL full outer
      join; sort keys lexicographically.
    * inner: use intersection of keys from both frames, similar to a SQL inner
      join; preserve the order of the left keys.
    * cross: creates the cartesian product from both frames, preserves the order
      of the left keys.

```python
pd.merge(data1,data2,how='left',on='key1')
```

```python
pd.merge(data1,data2,how='left',left_on='key2',right_on='key4')
```
