```python
b
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(4,3))
plt.hist(data,bins=10,color='#800000')
plt.xlabel('distinct datasets')
plt.ylabel('frequency')
plt.title('Frequency Graph')
plt.grid()
plt.savefig('hist.jpg')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
z
```

```python
plt.plot(x,y,z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='#800080',alpha=.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('3D Graph')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x,y,z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.bar(x,y,z)
```
